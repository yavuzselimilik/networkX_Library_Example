```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph():
    node_pos=nx.get_node_attributes(G,'pos')
    arc_weight=nx.get_edge_attributes(G,'weight')
    nx.draw_networkx(G, node_pos, node_size=450)
    nx.draw_networkx_edge_labels(G,node_pos,edge_labels=arc_weight,label_pos= 0.3)
    plt.axis('off')
    plt.show()
```

```python
G = nx.DiGraph()
```

```python
G.add_nodes_from([i for i in range(5)])
G.nodes()
```

```python
list_arcs = [(0, 1, 5.0), (0, 2, 3.0), (0, 4, 2.0), (1, 2, 2.0), (1, 3, 6.0), (2, 1, 1.0), (2, 3, 2.0), (4, 1, 6.0), (4, 2, 10.0), (4, 3, 4.0)]
```

```python
G.add_weighted_edges_from(list_arcs)
G.edges()
```

```python
G.nodes[0]['pos'] = (-2, 7)
G.nodes[1]['pos'] = (2, 3)
G.nodes[2]['pos'] = (1, -2)
G.nodes[3]['pos'] = (-5, -2)
G.nodes[4]['pos'] = (-6, 3)
```

```python
draw_graph()
```

```python
for x in range(5):
    try:
        print("4. köşeden ", x, ". köşeye en kısa yol: ",nx.dijkstra_path(G, source = 4, target=x))
    except:
        print("4. köşeden ", x, ". köşeye giden bir yol bulunamadı.")
```

```python
G.remove_node(3)
G.nodes()
```
